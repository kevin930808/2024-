```python
#4.58
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as st #計算z score面積

mean, std = 0, 1

x = np.linspace(mean-3*std, mean+3*std, 1000)


def main(a,b):
    return st.norm.cdf(b)-st.norm.cdf(a)
print('輸入計算之Z SCROE範圍A,B  P(A<Z<B)')
for i in range(3):
    A=float(input("A=>"))
    B=float(input("B=>"))
    ans=main(A,B)
    print("P(%.2f<Z<%.2f)=%.4f"%(A,B,ans))
    # 繪製常態分佈
    plt.plot(x, st.norm.pdf(x, mean, std))
    plt.axvline(A, color='red', linestyle='dashed', linewidth=1)
    plt.axvline(B, color='red', linestyle='dashed', linewidth=1)
    # 填充a和b之间的区域
    plt.fill_between(x, st.norm.pdf(x, mean, std), where=(A <= x) & (x <= B), color='green', alpha=0.5)
    plt.show()
```

```python
#4.86
from scipy.stats import gamma
import matplotlib.pyplot as plt
import numpy as np
shape, scale = 1.5, 4
rv = gamma(shape, scale=scale)

x = np.arange(rv.ppf(0.01), rv.ppf(0.99), 0.01)

y = rv.pdf(x)
plt.plot(x, y)
plt.title('Gamma distribution')
plt.xlabel('x')
plt.ylabel('Probability density')
plt.grid(True)
plt.show()
a_p= rv.cdf(3.5)
b_p=rv.cdf(1.75*2)#by w=0.5y
print('P(Y<3.5):', a_p)
print('P(W<1.75):', b_p)
```

```python
#4.128
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from scipy.optimize import root
#create probability density function f(x)=3*(1 - x)^2
def f(x):
    return 3 * (1 - x) ** 2
x = np.arange(0, 1, 0.01)
y = f(x)
plt.plot(x, y)
plt.fill_between(x, y, color='skyblue', alpha=0.4)
plt.title('Probability Density Function of f(x) = 3*(1 - x)^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# check f(x) 是否從0積分到1之面積(機率和)為1
# set F(x)為f(x)之積分
result, error = quad(f, 0, 1)
print("F(x) from 0 to 1 面積和:", result)


#create probability density function f(x)=3*(1 - x)^2
def func(x):
    return 3 * (1 - x) ** 2

# 積分區域 set from t 積分到1之解為0.1 by 4.128題目
def target_func(t):
    integral, _ = quad(func, t, 1)
    return integral - 0.1

#使用Root計算t值 
solution = root(target_func, 0)

print("t:", solution.x[0])
print("ans:",solution.x[0]*100)
```
